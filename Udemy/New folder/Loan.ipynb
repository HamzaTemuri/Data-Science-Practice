```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("credit_train.csv")
```

```python
data
```

```python
data.info()
```

```python
data.head()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
data['Years in current job'] = data['Years in current job'].map({'8 years':8,
                                                                 '10+ years':10,
                                                                 '3 years':3,
                                                                 '5 years':5,
                                                                 '< 1 year':0,
                                                                 '2 years':2,
                                                                 '4 years':4,
                                                                 '9 years':9,
                                                                 '7 years':7,
                                                                 '1 year':1,
                                                                 '6 years':6})
```

```python
data['Years in current job'].fillna((data['Years in current job'].mean()),inplace=True)
```

```python
data.drop('Months since last delinquent',axis=1,inplace=True)
```

```python
data['Credit Score'].fillna((data['Credit Score'].median()),inplace=True)
```

```python
data['Annual Income'].fillna((data['Annual Income'].median()),inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Loan Status'] = data['Loan Status'].map({'Fully Paid':1,'Charged Off':0})
```

```python
data['Term'] = data['Term'].map({'Short Term':0,'Long Term':1})
```

```python
data['Home Ownership'] = data['Home Ownership'].map({'HaveMortgage':0,'Home Mortgage':1,'Own Home':3,'Rent':4})
```

```python
data['Purpose'] = data['Purpose'].map({'Debt Consolidation':0,
                                       'Buy House':1,
                                       'other':2,
                                       'Business Loan':3,
                                       'Buy a Car':4,
                                       'major_purchase':5,
                                       'Take a Trip':6,
                                       'Other':7,
                                       'small_business':8,
                                       'Medical Bills':9,
                                       'wedding':10,
                                       'vacation':11,
                                       'Educational Expenses':12,
                                       'moving':13,
                                       'renewable_energy':14,
                                       'Home Improvements':15})
```

```python
data.drop(['Loan ID','Customer ID'],axis=1,inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('Loan Status',axis=1), data['Loan Status'], test_size=0.3, random_state=42)
```

```python
X_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=500)
rfc.fit(X_train,y_train)
fpred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,fpred))
print(classification_report(y_test,fpred))
```
